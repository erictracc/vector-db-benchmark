```python
import os
from pymongo import MongoClient
from dotenv import load_dotenv

# Load environment variables from .env
load_dotenv()

# Get MongoDB connection details from .env
mongo_uri = os.getenv("MONGO_URI")
mongo_db = os.getenv("MONGO_DB")
mongo_collection = os.getenv("MONGO_COLLECTION")

# Connect to MongoDB
client = MongoClient(mongo_uri)
db = client[mongo_db]
collection = db[mongo_collection]

# Insert a test document (this creates the DB and collection)
result = collection.insert_one({
    "message": "Hello, World! This is my first document using .env variables.",
    "test_id": 1
})

print(f"Inserted document with id: {result.inserted_id}")
```

```python
# ### CLEANUP CELL - Use this to reset your database
# Drops the entire 'documents' collection and all its data
collection.drop()

# Optional: Uncomment the line below to delete the entire 'vectordb' database
client.drop_database('vectordb')

print("Cleanup completed. Collection 'documents' has been dropped.")
print("Cleanup completed. Database 'vectordb' has been dropped.")
```
