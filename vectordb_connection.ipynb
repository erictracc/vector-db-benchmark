```python
import os
from qdrant_client import QdrantClient
from qdrant_client.http.models import VectorParams, Distance, PointStruct
from dotenv import load_dotenv

# Load environment variables from .env
load_dotenv()

# Get Qdrant connection details from .env
QDRANT_HOST = os.getenv("QDRANT_HOST")
QDRANT_PORT = int(os.getenv("QDRANT_PORT", 6333))
QDRANT_COLLECTION = os.getenv("QDRANT_COLLECTION", "documents")

# Connect to Qdrant
qdrant = QdrantClient(host=QDRANT_HOST, port=QDRANT_PORT)

# Check connection (list collections)
try:
    collections = qdrant.get_collections()
    print("✓ Connected to Qdrant!")
    print("Existing collections:", [c.name for c in collections.collections])
except Exception as e:
    print(f"✗ Connection error: {e}")
    exit(1)

# Create or recreate the collection
try:
    qdrant.recreate_collection(
        collection_name=QDRANT_COLLECTION,
        vectors_config=VectorParams(size=384, distance=Distance.COSINE)
    )
    print(f"✓ Collection '{QDRANT_COLLECTION}' created (or recreated).")
except Exception as e:
    print(f"✗ Error creating collection: {e}")

# Insert some sample vectors
points = [
    PointStruct(id=1, vector=[0.1]*384, payload={"content": "Hello Qdrant!"}),
    PointStruct(id=2, vector=[0.2]*384, payload={"content": "Another vector"})
]

try:
    qdrant.upsert(collection_name=QDRANT_COLLECTION, points=points)
    print("✓ Inserted points into Qdrant.")
except Exception as e:
    print(f"✗ Error inserting points: {e}")

# Perform a similarity search
try:
    search_result = qdrant.search(
        collection_name=QDRANT_COLLECTION,
        query_vector=[0.1]*384,
        limit=2
    )
    print("Search results:")
    for r in search_result:
        print(f"  → id={r.id}, score={r.score}, payload={r.payload}")
except Exception as e:
    print(f"✗ Error during search: {e}")
```
